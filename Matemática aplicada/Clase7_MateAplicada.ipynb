# Problema 3


Una empresa de tecnología quiere optimizar el tiempo de carga y descarga de archivos en su servicio de almacenamiento en la nube. Los tiempos medios para diferentes tamaños de archivos son los siguientes:

<p><center>

| Tamaño del archivo (MB) | Tiempo de carga (s) | Tiempo de descarga (s)|
|-------------------------|---------------------|-----------------------|
|$$50$$                   |$$10$$               |$$8$$                  |
|$$100$$                  |$$20$$               |$$16$$                 |
|$$250$$                  |$$50$$               |$$40$$                 |
|$$500$$                  |$$100$$              |$$80$$                 |
|$$1.000$$                 |$$200$$              |$$160$$                |

</center></p>

1. Encuentra las funciones que modelan el tiempo de carga $f(x)$ y el tiempo de descarga $g(x)$ en función del tamaño del archivo $x$. Utiliza la función *polyfit* de *Numpy* (debes encontrar dos funciones).
2. Define variable dependiente e independiente, para ambas funciones, indicando su unidad de medida.
3. Calcula el tiempo de carga y descarga para un archivo de $750$ MB.
4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño del archivo si el tiempo de carga es de $163$ segundos.
5. Un trabajador insiste en que si el tiempo de descarga es de $195$ segundos, el tamaño del archivo es $1.200$ MB. ¿Tiene razón? Utiliza *Python* para determinar.

```python
print(f"\nRespuesta 1:\nCalculamos con la formula de la pendiente para la funcion 1: (y2-y1)/(x2-x1)\n(20-10)/(100-50) = {(20-10)/(100-50)}, por lo tanto nuestra funcion 1 será:\nTc(tam) = tam*0.2\n");
print(f"Calculamos con la formula de la pendiente para la funcion 2: (y2-y1)/(x2-x1)\n(16-8)/(100-50) = {(16-8)/(100-50)}, por lo tanto nuestra funcion 2 será:\nTd(tam) = tam*0.16");
print(f"\nRespuesta 2: \nFuncion 1(Tc(tam)):\nVariable dependiente: Tiempo de carga en segundos Tc(tam) \nVariable independiente: Tamaño del archivo en MB (tam)\n");
print(f"Funcion 2(Td(Tam)):\nVariable dependiente: Tiempo de descarga en segundos Td(tam) \nVariable independiente: Tamaño del archivo en MB (tam)");
def Tc(tam):
    return tam*0.2
def Td(tam):
    return tam*0.16
def TamañoCarga(tam):
    return (tam*0.2) - 163
def TamañoDescarga(tam):
    return (tam*0.16) - 195
print(f"\nRespuesta 3:\nTiempo de carga para 750MB: {Tc(750)} segundos\nTiempo de descarga para 750MB: {Td(750)} segundos");
Tcarga = np.linspace(0,200,1)
re2 = fsolve(TamañoCarga,Tcarga)
print(f"\nRespuesta 4: El tamaño del archivo para 163 segundos de tiempo de carga será: {re2[0]:.0f} MB");
Tdescarga = np.linspace(0,200,1)
re2 = fsolve(TamañoDescarga,Tdescarga)
print(f"Respuesta 5: El tamaño del archivo para 195 segundos de tiempo de descarga será: {re2[0]:.0f} MB, por lo que la afirmación no es correcta pues el tamaño del archivo es un poco mas grande que 1.200 MB");
print(f"Respuesta 6: Grafico");
TimeCarga = np.arange(0,1000,1)
TimeDescarga = np.arange(0,1000,1)
plt.plot(TimeCarga,Tc(TimeCarga),label='Tiempo de carga', color='cyan')
plt.plot(TimeDescarga,Td(TimeDescarga),label='Tiempo de descarga', color='red')
plt.legend()
plt.title("Trafico de carga y descarga de archivos")
plt.xlabel("Tamaño del archivo (MB)")
plt.ylabel("Tiempo transcurrido (segundos)")
plt.grid(True)
plt.show()
print("")
```

# Problema 7

La depreciación es el proceso en que los activos físicos, como la maquinaria, vehículos, equipos, etc., pierden valor con el tiempo debido a factores como el desgaste, el uso y la obsolescencia. El valor estimado que tendrá el activo al final de su vida útil se conoce como *valor residual*.

En este contexto, sabemos que el valor residual $R(t)$ (en miles de dólares) de un equipo tecnológico está modelado por la función lineal:

$$R(t) = -1,2t + 10$$

donde $t$ corresponde a la antigüedad del equipo (en años).

7. Grafica la función utilizando la biblioteca *Matplotlib*. Interpreta lo observado.

```python
def R(t):
  return -1.2 * t + 10;
print(f"Respuesta 1: \nLa variable independiente es la antiguedad del equipo en años (t).\nLa variable dependiente es el valor residual en miles de dólares R(t)");
print("");
print("Respuesta 2: Representa el valor residual  del equipo respecto a la antiguedad de este último, en este caso es negativa por lo que podemos deducir que por cada año transcurrido el valor del equipo disminuira en 1.2");
print("");
print(f"Respuesta 3: \nEl valor residual inicial del equipo será {((R(0))*1000)} dólares");
print("");
def Rs(t):
  return (-1.2 * t + 10) - 0.4;
Rx = np.linspace(0,100,1);
re = fsolve(Rs,Rx);
print(f"Respuesta 4: \nPara saber la respuesta primero debemos sacar el valor final para la variable dependiente que será {re[0]:.0f}");
print("Por lo tanto el dominio será: [0;8] años")
print("");
print(f"Respuesta 5: \nEl valor residual del equipo tras 3 años y 6 meses será {((R(3.5))*1000)} dólares");
def Rs2(t):
  return (-1.2 * t + 10) - 4;
Rx = np.linspace(0,100,1);
re = fsolve(Rs2,Rx);
print("")
print(f"Respuesta 6: \nPara cuando el valor residual sea de 4000 dolares la antiguedad del equipo será de {re[0]:.0f} años");

Rt = np.arange(0,9,1)
plt.plot(Rt,R(Rt),label='Valor residual', color='cyan')
plt.legend()
plt.title("Valor residual de un equipo a través del tiempo")
plt.xlabel("Años transcurridos (años)")
plt.ylabel("Valor residual (Miles de dolares)")
plt.grid(True)
plt.show()
print("")
```

mientras que la distancia recorrida por el atleta $2$ está dada por la función:

4. Entre el segundo $8$ y el $10$, ¿cuál atleta iba más rápido? Justifique apropiadamente.

```python
def g(t):
  return 0.7 * (t**2)
def f(t):
  return 10*t
t1 = np.arange(0,11,1)
t2 = np.arange(0,11,1)
plt.plot(t1,g(t1),label='Atleta1', color='cyan')
plt.plot(t2,f(t2),label='Atleta2', color='red')
plt.legend()
plt.title("Distancia recorrida a traves del tiempo")
plt.xlabel("Tiempo transcurrido (segundos)")
plt.ylabel("Distancia recorrida (metros)")
plt.grid(True)
plt.show()
print("")
```
