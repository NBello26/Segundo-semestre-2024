```python
#Función para separar precios segun miles ejemplo: Tengo 10000 lo convertira en 10.000
def formatear_precio_miles(precio):
    return f"{precio:,.0f}".replace(',', 'X').replace('.', ',').replace('X', '.');

#np.log para logaritmos
#np.exp para exponentes

# Se llama a la biblioteca numpy, para trabajar con arreglos de datos
import numpy as np
# Módulo de la biblioteca scipy que nos permitirá poder resolver ecuaciones no necesariamente lineales
# a través de la función fsolve
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
import math
```

Sean las funciones:

\begin{split}
r(x) & = -1,5x^{2} + 11,5x - 15\\
g(x) & = 1,58x^{4} - 19,17x^{3} + 80,92x^{2} - 139,33x + 85
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $0 \leq x \leq 6$, $\hspace{0.3cm}$  $-10 \leq y \leq 10$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones considerando el dominio indicado. Sugerencia: utilice *plt.xlim()* y *plt.ylim()* para definir los límites de los ejes en el gráfico.
2. ¿Cuáles son los puntos de intersección entre las funciones? Utilice *fsolve* para encontrar las soluciones.

```python
def r(x):
    return (-1.5*(x**2)) + (11.5*x) - 15;

def g(x):
    return (1.58*(x**4)) - (19.17*(x**3)) + (80.92*(x**2)) - (139.33*(x)) + 85;

x = np.arange(0,15,0.1);

plt.plot(x,r(x), label = "r(x)");
plt.plot(x,g(x), label = "g(x)");

plt.xlim(0,6);
plt.ylim(-10,10);

plt.legend();
plt.title("Análisis de intercecciones");
plt.grid(True, color = "Aquamarine");
plt.show();

def rg(x):
    return r(x) - g(x);

x = np.linspace(0,15,10);
S = np.round(fsolve(rg,x),1);
S = np.unique(S);
print(f"Los valores de x son: \n1: {S[0]}\n2: {S[1]}\n3: {S[2]}\n4: {S[3]}");

print(f"Las intersecciones son: \n1: {S[0],round(g(S[0]),1)}\n2: {S[1],round(g(S[1]),1)}\n3: {S[2],round(g(S[2]),1)}\n4: {S[3],round(g(S[3]),1)}");

```

# Problema 2

Sean las funciones:

\begin{split}
h(x) & = 2(x - 2)^{2}\\
i(x) & = x + 1
\end{split}

<br>

1. Grafique ambas funciones en un mismo gráfico utilizando *Matplotlib*.
2. ¿En qué punto(s) interseca la función $h(x)$ al eje $x$? ¿y la función $i(x)$?
3. ¿Las gráficas de las funciones se intersecan en algún punto?
4. ¿Para qué valores de $x$ ocurre que $h(x) \leq i(x)$?
5. Determine el punto en que $h(x) = 50$.


```python
def h(x):
    return 2 * ((x-2) ** 2)
def i(x):
    return x+1

def hi(x):
    return h(x)-i(x)

x = np.arange(0,15,0.1);

plt.plot(x,h(x), label = "h(x)");
plt.plot(x,i(x), label = "i(x)");

plt.xlim(-3,6);
plt.ylim(-1,10);

plt.legend();
plt.title("Análisis de intercecciones");
plt.grid(True, color = "Aquamarine");
plt.show();






x = np.linspace(0,15,2);
S = np.round(fsolve(hi,x),1);
S = np.unique(S);
print(f"Los valores de x son: \n1: {S[0]}\n2: {S[1]}");

print(f"Las intersecciones son: \n1: {S[0],round(h(S[0]),1)}\n2: {S[1],round(h(S[1]),1)}");
```

# Problema 3

Sean las funciones:

\begin{split}
w(x) & = 0,6x^{2} + 1,5\\
t(x) & = 0,8x^{3} - 10x + 10
\end{split}

<br>

Considerando que $\hspace{0.2cm}$ $-3 \leq x \leq 4$, $\hspace{0.2cm}$ $-4 \leq y \leq 24$ $\hspace{0.2cm}$ realice lo siguiente:

1. Grafique ambas funciones en un mismo gráfico, indicando unidad de medida. Utilice *Matplotlib*.
2. Determine el intervalo en que $w(x) \geq t(x)$.
3. Encuentre los puntos en que $t(x) = 6$.


```python
def w(x):
    return (0.6*(x**2)) + 1.5
def t(x):
    return (0.8*(x**3)) - (10*x) + 10

x = np.arange(-3,4,0.1);

plt.plot(x,w(x), label = "w(x)");
plt.plot(x,t(x), label = "t(x)");

plt.xlim(-3,6);
plt.ylim(-5,15);

plt.legend();
plt.title("Análisis de intercecciones");
plt.grid(True, color = "Aquamarine");
plt.show();
```

```python
print("Respuesta 2:")

def z(x):
    return w(x) - t(x)
x = np.linspace(-6,6,3)

inter = np.round(fsolve(z,x),1)
inter_x = np.unique(inter)
print(inter_x)
print(f"Los intervalos donde w(x) >= t(x) son (-inf;-3,6] U [0,9;3,5]")
```

```python
print("Respuesta 3:")

def ts(x):
    return ((0.8*(x**3)) - (10*x) + 10) - 6

x = np.linspace(-6,6,3)


inter = np.round(fsolve(ts,x),1)
inter_x = np.unique(inter)
print(inter_x)






x = np.arange(-7,6,0.1);
plt.plot(x,w(x), label = "w(x)");
plt.plot(x,t(x), label = "t(x)");

plt.axhline(y = 6, color = "red") #Grafica una constante en el grafico
plt.xlim(-13,16);
plt.ylim(-5,25);

plt.legend();
plt.title("Análisis de intercecciones");
plt.grid(True, color = "Aquamarine");
plt.show();
```
