# Problema 3


Una empresa de tecnología quiere optimizar el tiempo de carga y descarga de archivos en su servicio de almacenamiento en la nube. Los tiempos medios para diferentes tamaños de archivos son los siguientes:

<p><center>

| Tamaño del archivo (MB) | Tiempo de carga (s) | Tiempo de descarga (s)|
|-------------------------|---------------------|-----------------------|
|$$50$$                   |$$10$$               |$$8$$                  |
|$$100$$                  |$$20$$               |$$16$$                 |
|$$250$$                  |$$50$$               |$$40$$                 |
|$$500$$                  |$$100$$              |$$80$$                 |
|$$1.000$$                 |$$200$$              |$$160$$                |

</center></p>

1. Encuentra las funciones que modelan el tiempo de carga $f(x)$ y el tiempo de descarga $g(x)$ en función del tamaño del archivo $x$. Utiliza la función *polyfit* de *Numpy* (debes encontrar dos funciones).
2. Define variable dependiente e independiente, para ambas funciones, indicando su unidad de medida.
3. Calcula el tiempo de carga y descarga para un archivo de $750$ MB.
4. Utiliza la función *fsolve* de *Scipy* para determinar el tamaño del archivo si el tiempo de carga es de $163$ segundos.
5. Un trabajador insiste en que si el tiempo de descarga es de $195$ segundos, el tamaño del archivo es $1.200$ MB. ¿Tiene razón? Utiliza *Python* para determinar.
6. Grafica ambas funciones en el mismo gráfico, indicando nombre de los ejes y unidades de medida. Utiliza la librería *Matplotlib* para graficar.

```python
print(f"\nRespuesta 1:\nCalculamos con la formula de la pendiente para la funcion 1: (y2-y1)/(x2-x1)\n(20-10)/(100-50) = {(20-10)/(100-50)}, por lo tanto nuestra funcion 1 será:\nTc(tam) = tam*0.2\n");
print(f"Calculamos con la formula de la pendiente para la funcion 2: (y2-y1)/(x2-x1)\n(16-8)/(100-50) = {(16-8)/(100-50)}, por lo tanto nuestra funcion 2 será:\nTd(tam) = tam*0.16");
print(f"\nRespuesta 2: \nFuncion 1(Tc(tam)):\nVariable dependiente: Tiempo de carga en segundos Tc(tam) \nVariable independiente: Tamaño del archivo en MB (tam)\n");
print(f"Funcion 2(Td(Tam)):\nVariable dependiente: Tiempo de descarga en segundos Td(tam) \nVariable independiente: Tamaño del archivo en MB (tam)");
def Tc(tam):
    return tam*0.2
def Td(tam):
    return tam*0.16
def TamañoCarga(tam):
    return (tam*0.2) - 163
def TamañoDescarga(tam):
    return (tam*0.16) - 195
print(f"\nRespuesta 3:\nTiempo de carga para 750MB: {Tc(750)} segundos\nTiempo de descarga para 750MB: {Td(750)} segundos");
Tcarga = np.linspace(0,200,1)
re2 = fsolve(TamañoCarga,Tcarga)
print(f"\nRespuesta 4: El tamaño del archivo para 163 segundos de tiempo de carga será: {re2[0]:.0f} MB");
Tdescarga = np.linspace(0,200,1)
re2 = fsolve(TamañoDescarga,Tdescarga)
print(f"Respuesta 5: El tamaño del archivo para 195 segundos de tiempo de descarga será: {re2[0]:.0f} MB, por lo que la afirmación no es correcta pues el tamaño del archivo es un poco mas grande que 1.200 MB");
print(f"Respuesta 6: Grafico");
TimeCarga = np.arange(0,1000,1)
TimeDescarga = np.arange(0,1000,1)
plt.plot(TimeCarga,Tc(TimeCarga),label='Tiempo de carga', color='cyan')
plt.plot(TimeDescarga,Td(TimeDescarga),label='Tiempo de descarga', color='red')
plt.legend()
plt.title("Trafico de carga y descarga de archivos")
plt.xlabel("Tamaño del archivo (MB)")
plt.ylabel("Tiempo transcurrido (segundos)")
plt.grid(True)
plt.show()
print("")
```
